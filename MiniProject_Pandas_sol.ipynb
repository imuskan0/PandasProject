```python
#Identify and separate the columns representing homework scores and their respective maximum points.
homework_scores = []
homework_max_points = []

for col in final_df.columns:
    if 'Homework' in col:
        if 'Max Points' in col:
            homework_max_points.append(col)
        else:
            homework_scores.append(col)

homework_scores, homework_max_points
```

**Question 5: Calculating Homework Score by Average Method**

```python
total_ratio_sum = 0

for i in range(len(homework_scores)):
    ratio = final_df[homework_scores[i]] / final_df[homework_max_points[i]]
    total_ratio_sum += ratio

average_Homework_Score = total_ratio_sum / len(homework_scores)

```

**Question 6: Storing Homework Score by Average Method**

```python
#Add the calculated average homework scores to the DataFrame.
final_df['Average Homework Score'] = average_Homework_Score
final_df[['Average Homework Score']].head()
```

```python
#Determine the final homework score for each student by taking the maximum value between the total and average methods.
final_df['Final Homework Score'] = final_df[['Homework Score', 'Average Homework Score']].max(axis=1)
final_df.head()
```

## Group the data to calculate final scores!

```python
# weights = pd.Series(
#     {
#         "Exam 1 Score": 0.05,
#         "Exam 2 Score": 0.10,
#         "Exam 3 Score": 0.15,
#         "Quiz Score": 0.30,
#         "Homework Score": 0.40,
#     }
# )
```

1. Define the weights for each component.
2. Calculate the final score by applying the weights to each component and summing the results.
3. Calculate the ceiling score by rounding up the final score multiplied by 100.
4. Define a function to convert numeric scores to letter grades based on standard thresholds.
5. Map the ceiling scores to letter grades.
6. Store the letter grades as a categorical column in the DataFrame.
7. Display the final letter grades.

```python
#1
weights = pd.Series({
    "Exam 1 Score": 0.05,
    "Exam 2 Score": 0.10,
    "Exam 3 Score": 0.15,
    "Quiz Score": 0.30,
    "Homework Score": 0.40,
})

#2
final_df['Final Score'] = (
    final_df['Exam 1 Score'] * weights['Exam 1 Score'] +
    final_df['Exam 2 Score'] * weights['Exam 2 Score'] +
    final_df['Exam 3 Score'] * weights['Exam 3 Score'] +
    final_df['Quiz Score'] * weights['Quiz Score'] +
    final_df['Homework Score'] * weights['Homework Score']
)

#3
final_df['Ceiling Score'] = np.ceil(final_df['Final Score'] * 100)

#4
def score_to_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

#5
final_df['Final Grade'] = final_df['Ceiling Score'].apply(score_to_grade)

#6
final_df[['Email Address', 'Final Score', 'Ceiling Score', 'Final Grade']].head()
```
